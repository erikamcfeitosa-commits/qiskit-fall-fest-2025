## Conceitos Quânticos Aplicados

### 1. Quibit e Superposição
O qubit é a unidade fundamental da informação quântica.
No jogo, ele inicia no estado |0⟩ e, ao aplicar a porta Hadamard (H), entra em superposição — passando a representar simultaneamente os estados |0⟩ e |1⟩.
Essa etapa é visualizada na esfera de Bloch, que representa o qubit entre os polos clássico de 0 e 1.

### 2. Portas Quânticas e Transformações de Estado
As portas quânticas funcionam como operações lógicas que alteram o estado do qubit.
No jogo, o participante pode aplicar:
Porta X: inverte o estado.
Porta H (Hadamard): coloca o qubit em superposição.
Essas transformações são imediatamente refletidas no circuito quântico exibido no notebook.

### 3. Medição e Colapso da Função de Onda
Ao medir o qubit, ocorre o colapso da função de onda — o estado deixa de ser uma combinação probabilística e assume um valor clássico (|0⟩ ou |1⟩).
A simulação no Qiskit mostra esse colapso de forma probabilística, de acordo com o histórico de portas aplicadas.

### 4. Interatividade
O notebook utiliza widgets interativos (botões) para o jogador escolher ações: aplicar portas (X, H) ou medir.  
Cada escolha altera o circuito e a visualização, criando um aprendizado ativo e engajante e ajudado o jogador a construir intuição visual sobre o comportamento quântico — algo que normalmente é ensinado apenas com matemática complexa.

Este bloco importa todas as classes e funções essenciais do Qiskit, Qiskit Aer, Matplotlib e Ipywidgets. Ele também inicializa o simulador `aer_simulator` do Qiskit Aer, que será usado para executar os circuitos quânticos.

```python
# --- Importações principais ---
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
import matplotlib.pyplot as plt

# --- Inicializa simulador ---
backend = Aer.get_backend('aer_simulator')
```

Esta célula exibe uma introdução visual e narrativa para o jogo, usando HTML para formatar o texto e adicionar um título estilizado. Ela prepara o usuário para a experiência interativa.

```python
# --- Introdução visual e narrativa ---
display(HTML("""
<h1 style='color:#7e57c2;text-align:center'>🌀 2025: Uma Odisseia no Mundo Quântico </h1>
<p style='text-align:center;max-width:600px;margin:auto'>
Você é um explorador quântico, navegando entre a incerteza e o controle.<br>
Cada porta aplicada altera o estado do seu qubit, mudando a probabilidade de ele estar em |0⟩ ou |1⟩. No mundo quântico, o qubit pode existir em superposição — uma combinação entre 0 e 1 ao mesmo tempo.
Mas quando o observamos, essa superposição colapsa, revelando apenas um dos resultados.<br><br>
<strong>Use as portas H, X, Y e Z para manipular o qubit e descobrir o que acontece.</strong><br>
Quando estiver pronto, clique em <strong>Medir</strong> para colapsar o universo.
</p>
"""))
```

Aqui, um circuito quântico de 1 qubit (`qc = QuantumCircuit(1)`) é criado. A função `atualizar_visualizacao` é definida para limpar a saída anterior, exibir a explicação atual, o circuito quântico e a Esfera de Bloch. Ela garante que a visualização seja sempre limpa e atualizada.

```python
# --- Criação do circuito quântico ---
qc = QuantumCircuit(1)

# --- Função para atualizar visualização ---
def atualizar_visualizacao(qc, explicacao):
    """Atualiza visualização do circuito e estado na esfera de Bloch."""
    # Limpa a saída anterior para garantir uma nova renderização limpa
    clear_output(wait=True)

    display(HTML(f"<h3 style='color:#5e35b1'>{explicacao}</h3>"))
    display(qc.draw('mpl'))

    # Obter vetor de estado de forma robusta
    # Criar uma cópia temporária do circuito e adicionar save_statevector para garantir o resultado
    qc_temp = qc.copy()
    qc_temp.save_statevector()
    state = backend.run(transpile(qc_temp, backend)).result().get_statevector()

    # Plotar esfera de Bloch
    fig_bloch = plot_bloch_multivector(state, title="Esfera de Bloch — Estado Atual")
    display(fig_bloch)
    plt.close(fig_bloch) # Fecha a figura para liberar recursos
```

Esta função `aplicar_porta` é o coração da interatividade. Ela recebe o nome de uma porta (H, X, Y, Z, Medir, Reset) e aplica a operação correspondente ao qubit. Em caso de 'Medir', ela executa o circuito no simulador, obtém os resultados e plota um histograma das probabilidades. Após cada ação, ela atualiza a visualização e reexibe os botões.

```python
def aplicar_porta(porta):
    """Aplica a porta escolhida e atualiza visualmente o circuito."""
    global qc, botoes # Adicione 'botoes' como global para garantir acesso
    if porta == 'H':
        qc.h(0)
        explic = "Porta Hadamard aplicada — o qubit agora está em superposição."
    elif porta == 'X':
        qc.x(0)
        explic = "Porta X aplicada — invertendo o estado 0↔1."
    elif porta == 'Z':
        qc.z(0)
        explic = "Porta Z aplicada — alterando a fase do qubit."
    elif porta == 'Y':
        qc.y(0)
        explic = "Porta Y aplicada — rotação combinada entre X e Z."
    elif porta == 'Medir':
        qc.measure_all()
        # Correção: remove assemble(), que foi descontinuado e não é mais necessário.
        result = backend.run(transpile(qc, backend), shots=1024).result()
        counts = result.get_counts()
        display(HTML("<h3 style='color:#43a047'>🔮 O colapso ocorreu! Resultado:</h3>"))

        # Exibir o histograma explicitamente e fechar a figura
        fig_hist = plot_histogram(counts)
        display(fig_hist)
        plt.close(fig_hist) # Fecha a figura para liberar recursos

        # Mostrar explicação educativa
        display(HTML("""
        <p style='font-size:16px;max-width:600px;margin:auto;text-align:center'>
        Cada medição é única — você acabou de testemunhar o <strong>colapso da função de onda</strong>.<br>
        O resultado não é determinado pelas portas, mas pelas <em>probabilidades</em> que elas criaram.<br>
        Bem-vindo à fronteira da incerteza quântica.
        </p>
        """))
        return
    elif porta == 'Reset':
        qc = QuantumCircuit(1)
        explic = "Circuito resetado — o qubit retornou ao estado |0⟩."

    atualizar_visualizacao(qc, explic)
    # Reexibir os botões após cada ação para que não desapareçam
    display(HTML("<h3 style='color:#6a1b9a'>Escolha sua próxima ação quântica:</h3>"))
    display(widgets.HBox(botoes))
```

Este bloco cria os botões interativos (`H`, `X`, `Y`, `Z`, `Medir`, `Reset`) usando `ipywidgets.Button`. Ele atribui a função `aplicar_porta` a cada botão, para que, ao serem clicados, a ação correspondente seja executada. Finalmente, ele inicializa a visualização com o qubit no estado `|0⟩` e exibe os botões para o usuário interagir.

```python
# --- Botões do jogo ---
botoes = [widgets.Button(description=p, button_style='') for p in ['H','X','Y','Z','Medir', 'Reset']]

# --- Atribuir comportamento ---
for b in botoes:
    b.on_click(lambda x, b=b: aplicar_porta(b.description))

# --- Estado inicial ---
atualizar_visualizacao(qc, "O qubit inicia em |0⟩ — a origem da sua jornada.")

# --- Exibir controles (inicialmente) ---
# Eles foram limpos pelo atualizar_visualizacao acima, então os reexibimos.
display(HTML("<h3 style='color:#6a1b9a'>Escolha sua próxima ação quântica:</h3>"))
display(widgets.HBox(botoes))
```
